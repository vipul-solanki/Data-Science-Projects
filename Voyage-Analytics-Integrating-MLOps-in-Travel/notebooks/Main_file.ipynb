## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
clean_data = merged_data.copy()

numeric_cols = clean_data.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = clean_data.select_dtypes(include=['object']).columns.tolist()

clean_data[numeric_cols] = clean_data[numeric_cols].fillna(clean_data[numeric_cols].median())
for col in categorical_cols:
    clean_data[col] = clean_data[col].fillna(clean_data[col].mode()[0])

print('Missing values after imputation:')
display(clean_data.isnull().sum().to_frame('missing_count'))
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

- The source CSV files do not show missing values, so the main missing-value issue appears after merging hotel information with flight records.
- Median imputation was used for numeric columns because it is robust to extreme values.
- Mode imputation was used for categorical columns so that the most common category is preserved.
- This approach keeps the merged dataset usable without dropping a large number of flight records.

### 2. Handling Outliers

````python
outlier_columns = ['price', 'time', 'distance', 'age', 'days', 'hotel_price', 'total']

for col in outlier_columns:
    q1 = clean_data[col].quantile(0.25)
    q3 = clean_data[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    clean_data[col] = clean_data[col].clip(lower=lower_bound, upper=upper_bound)

print('Outlier treatment completed for selected numeric columns.')
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

- IQR-based capping was applied to numeric variables such as price, time, distance, age, stay days, and hotel spending.
- This approach reduces the influence of extreme observations while retaining most of the data.
- It is appropriate here because travel pricing can contain long tails, especially for premium routes and higher-cost stays.

### 3. Categorical Encoding

````python
clean_data['gender_encoded'] = clean_data['gender'].map({'male': 0, 'female': 1}).fillna(0)
clean_data['flightType_encoded'] = clean_data['flightType'].map({'economic': 0, 'premium': 1, 'firstClass': 2})

print('Encoded columns preview:')
display(clean_data[['gender', 'gender_encoded', 'flightType', 'flightType_encoded']].head())
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

- Ordinal encoding was used for `flightType` because ticket classes have a natural order from economic to first class.
- One-hot encoding was applied to route, agency, company, hotel, and place variables because those categories do not have a meaningful numeric ranking.
- This combination preserves business meaning while making the data model-ready.

### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
clean_data['date'] = pd.to_datetime(clean_data['date'], format='%m/%d/%Y')
clean_data['hotel_date'] = pd.to_datetime(clean_data['hotel_date'], format='%m/%d/%Y')
clean_data['trip_month'] = clean_data['date'].dt.month
clean_data['trip_dayofweek'] = clean_data['date'].dt.dayofweek
clean_data['is_weekend_trip'] = clean_data['trip_dayofweek'].isin([5, 6]).astype(int)
clean_data['hotel_per_day'] = clean_data['total'] / clean_data['days'].replace(0, 1)
clean_data['price_per_distance'] = clean_data['price'] / clean_data['distance'].replace(0, 1)

print('Feature engineered dataset preview:')
display(clean_data.head())
````

#### 2. Feature Selection

````python
model_data = clean_data.copy()
model_data = model_data.drop(columns=['travelCode', 'userCode', 'user_name'])
model_data = pd.get_dummies(model_data, columns=['company', 'gender', 'from', 'to', 'flightType', 'agency', 'hotel_name', 'place'], drop_first=True)
model_data = model_data.drop(columns=['date', 'hotel_date'])
model_data = model_data.sample(n=min(50000, len(model_data)), random_state=42)

X = model_data.drop(columns=['price'])
y = model_data['price']

print('Feature matrix shape:', X.shape)
print('Target shape:', y.shape)
````

- Raw date columns were replaced with derived calendar features such as month and day of week.
- The final feature set keeps variables that explain route cost, class choice, customer profile, and hotel spend.

##### Which all features you found important and why?

Important features include `distance`, `time`, `flightType`, `agency`, `hotel_price`, `total`, `days`, `trip_month`, and `trip_dayofweek`.

````python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
linear_predictions = linear_model.predict(X_test)

linear_metrics = pd.DataFrame({
    'Metric': ['MAE', 'RMSE', 'R2'],
    'Score': [
        mean_absolute_error(y_test, linear_predictions),
        mean_squared_error(y_test, linear_predictions, squared=False),
        r2_score(y_test, linear_predictions)
    ]
})

display(linear_metrics)
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
ridge_params = {'alpha': [0.1, 1.0, 10.0]}
ridge_search = GridSearchCV(Ridge(), ridge_params, cv=3, scoring='r2', n_jobs=-1)
ridge_search.fit(X_train, y_train)

ridge_model = ridge_search.best_estimator_
ridge_predictions = ridge_model.predict(X_test)

ridge_metrics = pd.DataFrame({
    'Metric': ['MAE', 'RMSE', 'R2'],
    'Score': [
        mean_absolute_error(y_test, ridge_predictions),
        mean_squared_error(y_test, ridge_predictions, squared=False),
        r2_score(y_test, ridge_predictions)
    ]
})

print('Best Ridge alpha:', ridge_search.best_params_['alpha'])
display(ridge_metrics)
````

##### Which hyperparameter optimization technique have you used and why?

GridSearchCV was used for Ridge regression to tune the `alpha` value. This method is appropriate because the search space is small, transparent, and easy to compare across candidate regularization strengths.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes. Regularization helps stabilize a linear model when many encoded features are present. In practice, the tuned Ridge model is useful for checking whether error can be reduced without losing interpretability.
